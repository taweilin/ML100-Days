```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cv2.cvtColor is an OpenCV function to convert images to different color spaces
def convertToRGB(img): 
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#load test iamge
test1 = cv2.imread('data/test5.JPG')
#convert the test image to gray image as opencv face detector expects gray images 
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
```

## 1.

```python
#load cascade classifier training file for haarcascade 
haar_face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_alt.xml')
```

```python
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.3, minNeighbors=1);  

#go over list of faces and draw them as rectangles on original colored 
for (x, y, w, h) in faces:     
         cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2)
        #let's detect multiscale (some images may be closer to camera than others) images 
#print the number of faces found 
print('Faces found: ', len(faces))
```

```python
#convert image to RGB and show image 
plt.imshow(convertToRGB(test1))
```

# 2.

```python
#load test iamge
test2 = cv2.imread('data/test5.JPG')
```

```python
def detect_faces(f_cascade, colored_img, scaleFactor = 1.1):
 #just making a copy of image passed, so that passed image is not changed 
 img_copy = colored_img.copy()          
 
 #convert the test image to gray image as opencv face detector expects gray images
 gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)          
 
 #let's detect multiscale (some images may be closer to camera than others) images
 faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=1);          
 
 #go over list of faces and draw them as rectangles on original colored img
 for (x, y, w, h) in faces:
      cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)              
 
 return img_copy
```

```python
faces2 = detect_faces(haar_face_cascade, test2);  
print('Faces found: ', len(faces2))
```

```python
#convert image to RGB and show image 
plt.imshow(convertToRGB(faces2))
```

# 3.

```python
#load cascade classifier training file for haarcascade 
lbp_face_cascade = cv2.CascadeClassifier('data/lbpcascade_frontalface.xml')
```

```python
faces3 = detect_faces(lbp_face_cascade, test2); 

print('Faces found: ', len(faces3))
```

```python
#convert image to RGB and show image 
plt.imshow(convertToRGB(faces3))
```
